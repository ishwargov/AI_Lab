$f1 = x^2 + 2x + 1$ 

$f2 = x^2 - 2x + 1$


```
def f1(x):
  ret = x**2 + 2*x + 1
  return ret

def f2(x):
  ret = x**2 - 2*x + 1
  return ret
```

```
x = np.linspace(-5,5,100)
```

```
plt.plot(x,f1(x),c='r',label='f1')
plt.plot(x,f2(x),c='b',label='f2')
plt.xlabel('X')
plt.ylabel('f1 and f2')
plt.legend()
```

```
import numpy as np
import re
from collections import Counter
```

```
f = open('speechs.txt')
dat = f.read()
```

```
#dat = re.sub('"'," ",dat)
#dat = re.sub("'"," ",dat)
```

```
dat_words = dat.split()
words = {x:dat_words.count(x) for x in dat_words}
```

```
i = 0
word_idx = dict()
idx_word = dict()
for word in words:
  word_idx[word] = i
  idx_word[i] = word 
  i+=1
```

```
word_idx['super'],idx_word[1986]
```

```
print("Total unique words = ",len(words))
```

```
cnt_data = np.zeros(shape=(len(words),len(words)))     #frequency table 2d array
```

```
for i in range(len(dat_words)-1):
  cnt_data[word_idx[dat_words[i]]][word_idx[dat_words[i+1]]]+=1
cnt_data
```

```
freq_data = np.array(cnt_data)
```

```
for i in range(len(words)):
  if(sum(freq_data[i,:])!=0):
    freq_data[i,:] = freq_data[i,:]/sum(freq_data[i,:])
```

```
freq_data
```

```
cnt_data
```

```
def gen_next_word(word): 
  idx = word_idx[word]
  rng = np.random.rand()
  arr = []
  for i in range(len(words)):
    arr.append((freq_data[idx,i],i))
  arr.sort()
  arr.reverse()
  cur = 0
  j = 0
  while(rng>cur+arr[j][0] and j<len(words)):
    cur += arr[j][0]
    j+=1
  if(j>len(words)):
    j-=1
  return(idx_word[arr[j][1]])
```

```
def freq_next_word(word):
  idx = word_idx[word]
  return(cnt_data[idx,:])
```

```
freq_list = freq_next_word('one').tolist()
i = 0
for j in freq_list:
  print('(',idx_word[i],j,')',end='')
  i+=1
```

```
gen_sample = []
```

```
nparas = np.random.randint(25,50)
for _ in range(nparas):
  nwords = int(abs(np.random.normal(25,5)))
  rng = np.random.randint(0,len(words))
  cur_word = idx_word[rng]
  print(cur_word,end=' ')
  for i in range(nwords):
    gen_sample.append(cur_word)
    cur_word = gen_next_word(cur_word)
    print(cur_word,end=' ')
  print()
```

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##Q1

```
def k_means(data,k,centre):
  dat = np.array(data)
  np.random.shuffle(dat)
  centre = np.empty((0,2))
  if(len(centre)==0):
    for i in range(k):
      centre = np.append(centre,[dat[i]],axis=0)
  while(1):
    dist_val = np.empty((0,len(dat)))
    for i in range(k):
      dist_val = np.append(dist_val,[np.linalg.norm(dat-centre[i],axis=1)],axis=0)
    grp = np.argmin(dist_val,axis=0)
    cnt = np.zeros((k,1))
    centroid = np.zeros((k,2))
    sq_error = 0
    for i in range(len(grp)):
      centroid[grp[i]] += [dat[i,0],dat[i,1]]
      cnt[grp[i]] += 1
      sq_error += np.linalg.norm(centre[grp[i]]-dat[i])
    for i in range(k):
      if(cnt[i]!=0):
        centroid[i] = centroid[i]/cnt[i]
      else:
        centroid[i] = centre[i]
    if(np.array_equal(centre,centroid)):
      break
    else:
      centre = np.array(centroid)
  return((dat,grp,centre,sq_error))
```

```
dat = np.append(np.random.normal(5,0.5,(1000,2)),np.append(np.random.normal(2,0.5,(1000,2)),np.random.normal(-2,0.5,(1000,2)),axis=0),axis=0)
k = 3
cluster = k_means(dat,k,None)
plt.figure(figsize=(7,7))
plt.scatter(cluster[0][:,0],cluster[0][:,1],c=cluster[1],ec='black')
plt.scatter(cluster[2][:,0],cluster[2][:,1],c="red",marker='*')
print("Sum of Squared Errors :",cluster[3])
```

```python
import numpy as np
```

```python
p_out = np.zeros((10,5))
p_out_min = [0.01,0.02,0.03,0.1,0.3]
p_out_max = [0.1,0.2,0.3,0.5,0.7]
a = np.array([1,2,3,4,6])
```

```python
for i in range(10):
  for j in range(5):
    p_out[i,j] = p_out_max[j] + (p_out_min[j]-p_out_max[j])/9 * ((i+1)-1)
```

```python
p_out
```

```python
best_score = np.zeros((300,10))
best_shot = np.zeros((300,10))
```

```python
for w in range(10):
  if(w>=9):
    best_score[0,w] = max((1-p_out[w,:])*a)
    best_shot[0,w] = a[np.argmax((1-p_out[w,:])*a,axis=0)]
```

```python
for w in range(10):
  for b in range(1,300):
    if(b+w>=9):
      best_score[b,w] = max( (1-p_out[w,:])*(a+best_score[b-1,w]) + (p_out[w,:]*best_score[b-1,w-1] if (w>0) else 0)) 
      best_shot[b,w] = a[np.argmax( (1-p_out[w,:])*(a+best_score[b-1,w]) + (p_out[w,:]*best_score[b-1,w-1] if (w>0) else 0))]
```
